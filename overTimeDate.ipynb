```python
#first import the excel sheet
import pandas as pd
import plotly.express as px
from plotly import graph_objects as go
import ipywidgets as widgets
from IPython.display import display

import numpy as np
import spacy
from collections import Counter

import dash
from dash import dcc, html
from dash.dependencies import Input, Output

import dash_bootstrap_components as dbc #used for formatting layout of the dash

file_path = "C:/Users/rgae/OneDrive - QuidelOrtho/Documents/All QuidelFiles/Excel Files/6-27-2024 2nd Copy .xlsx"
df = pd.read_excel(file_path, usecols = 'AN, AP, AV, AZ')
```

```python
#Filter the DataFrame
df_filtered_AZ = df.loc[df['Action Taken Description'] == 'CLOSE A CALL'].copy().reset_index()

df_filtered_AZ['Incident Close Loc Dt'] = pd.to_datetime(df_filtered_AZ['Incident Close Loc Dt'])
df_filtered_AZ['Incident Creation Loc Dt'] = pd.to_datetime(df_filtered_AZ['Incident Creation Loc Dt'])

df_filtered_AZ['Difference'] = df_filtered_AZ['Incident Close Loc Dt'] - df_filtered_AZ['Incident Creation Loc Dt']
df_filtered_AZ['Difference in Days'] = df_filtered_AZ['Difference'].dt.days
df_filtered_AZ = df_filtered_AZ.drop(columns = ['Difference', 'Action Taken Description', 'Problem Dsc', 'index'])

df_filtered_AZ['exist'] = 1

# Save the resulting DataFrame to an Excel file
df_filtered_AZ.to_excel('difference_date.xlsx')
```

```python
fig = px.histogram(df_filtered_AZ, x = 'Incident Creation Loc Dt', y = 'exist', title = 'Difference Over Time')
fig.show()
```
